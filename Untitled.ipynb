```python
import numpy as np
import matplotlib.pyplot as plt

def local_regression(x0,X,Y,tau):
    x0 = [1,x0]
    X = [[1,i] for i in X]
    X = np.asarray(X)
    xw = (X.T)*np.exp(np.sum((X-x0)**2,axis = 1)/(-2*tau))
    beta = np.linalg.pinv(xw @ X)@ xw @Y @x0
    return beta
def draw(tau):
    prediction = [local_regression(x0,X,Y,tau) for x0 in domain]
    plt.plot(X,Y,'o',color='black')
    plt.plot(domain,prediction,color='red')
    plt.show()
    
    
    
X = np.linspace(-3,3,num=1000)
domain = X
Y = np.log(np.abs(X**2 -1)+0.5)
draw(10)
draw(1)
draw(0.1)
draw(0.01)
```

```python
from sklearn import model_selection,neighbors,datasets

iris = datasets.load_iris()
x_train,x_test,y_train,y_test = model_selection.train_test_split(iris.data,iris.target,test_size = 10)
# print(iris)
for i in range(len(iris.target_names)):
    print(f'label {i}','-',str(iris.target_names[i]))

classifier = neighbors.KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train,y_train)
pred = classifier.predict(x_test)

for i in range(len(x_test)):
    print(f'sample: {x_test[i]} actual-label:{y_test[i]} predicted-label:{pred[i]}')
    
print(f'accuracy:{classifier.score(x_test,y_test)}')
```

```python
import matplotlib.pyplot as plt
from sklearn import cluster, datasets, mixture, preprocessing
import numpy as np

dataset = datasets.load_iris()
X = preprocessing.StandardScaler().fit_transform(dataset.data)
models = [None, cluster.KMeans(n_clusters=3), mixture.GaussianMixture(n_components=3)]

graph_names = ['Real', 'Kmeans', 'GaussianMixture']

plt.figure(figsize=(14, 7))

for i, model in enumerate(models):
    plt.subplot(1, 3, i+1)
    if model:
        labels = model.fit_predict(X)
    else:
        labels = dataset.target
    plt.scatter(X[:, 2], X[:, 3], c=labels)  # You can choose a different colormap
    plt.title(graph_names[i])

plt.show()
```

```python
# candidate elimination
import numpy as np
import pandas as pd

df = pd.read_csv('/home/prax/Downloads/sports4.csv')
df = df.drop(['slno'],axis = 1)
concept = df.values[:,:-1]
target = df.values[:,-1]

def learn(concepts,target):
    specific = concepts[0].copy()
#     general = [["?" for _ in range(len(specific))] for _ in range(len(specific))]
    general = [["?" for i in range(len(specific))] for i in range(len(specific))]
    for j,example in enumerate(concepts):
        if target[j] == 'yes':
            for i in range(len(specific)):
                if example[i]!=specific[i]:
                    specific[i] = "?"
                    general[i][i] = "?"
        elif target[j] == "no":
            for i in range(len(specific)):
                if example[i]!=specific[i]:
                    general[i][i] = specific[i]
                else:
                    general[i][i] = "?"

    ind = [i for i,val in enumerate(general) if val == ['?','?','?','?','?','?']]
    for i in ind:
        general.remove(['?','?','?','?','?','?'])
    return specific,general
    
specific,general=learn(concept,target)
print(f'S:{specific}')      
print(f'G:{general}')
```

```python
# id3 algorithm

import pandas as pd
from sklearn.feature_selection import mutual_info_classif
from pprint import pprint
from collections import Counter

def id3(data,target,attribute):
    cnt = Counter(x for x in data[target])
    if len(cnt) == 1:
        return next(iter(cnt))
    elif data.empty or (not attribute):
        return None
    else:
        gainz = mutual_info_classif(data[attribute],data[target],discrete_features=True)
        indexof = gainz.tolist().index(max(gainz))
        best = attribute[indexof]
        tree = {best:{}}
        rem = [i for i in attribute if i!=best]
        for attval,dataset in data.groupby(best):
            subtree = id3(dataset,target,rem)
            tree[best][attval] = subtree
        return tree
    
data = pd.read_csv('/home/prax/Downloads/tennis.csv')
attr = data.columns.tolist()
attr.remove('PlayTennis')

for col in data.select_dtypes('object'):
    data[col],_= data[col].factorize()
    
tree = id3(data,'PlayTennis',attr)
pprint(tree)
```
